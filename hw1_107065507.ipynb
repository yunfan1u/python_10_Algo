## Implementations and comparison of five sorting methods:
    * Selection Sort
    * Bubble Sort
    * Merge Sort
    * Heap Sort
    * Quick Sort
    
## Time complexity:
    * Selection Sort: O(n^2)
    * Bubble Sort: O(n^2)
    * Merge Sort: O(nlgn)
    * Heap Sort: O(nlgn)
    * Quick Sort: O(nlgn) (in worst case: O(n^2)

```python
import random
import time
class hw_list(list):
    
    def SelectSort(self):
        for i in range(len(self)-1):
            for j in range(i+1, len(self)):
                if self[i] > self[j]:
                    self[i], self[j] = self[j], self[i]
        return
    
    def BubbleSort(self):
        for i in range(len(self)):  
            for j in range(0, len(self)-i-1): 
                if self[j] > self[j+1] : 
                    self[j], self[j+1] = self[j+1], self[j]
        return
    
    def MergeSort(self):
        if len(self)<=1:
            return
        else:
            mid = len(self)//2
            left = hw_list(self[:mid])
            right = hw_list(self[mid:])
            
            left.MergeSort()
            right.MergeSort()
            
            i=0
            j=0
            k=0
            
            while i < len(left) and j < len(right):
                if left[i] < right[j]:
                    self[k]=left[i]
                    i=i+1
                else:
                    self[k]=right[j]
                    j=j+1
                k=k+1

            while i < len(left):
                self[k]=left[i]
                i=i+1
                k=k+1

            while j < len(right):
                self[k]=right[j]
                j=j+1
                k=k+1
        return
    
    #HeapSort
    def shiftup(self, i):
        if i == 0:
            return
        else:
            if self[i] <= self[(i-1)//2]:
                return
            else:
                self[i], self[(i-1)//2] = self[(i-1)//2], self[i]
                self.shiftup((i-1)//2)
                
    def heapify(self):    # build heap??
        for i in range(len(self)):  #Top-Down
            self.shiftup(i)
        return
    
    def shiftdown(self, i, end):    
        lc = 2*i+1
        rc = 2*i+2
        if lc > end:
            return
        elif lc == end:
            if self[i] < self[lc]:
                self[i], self[lc] = self[lc], self[i]
            return
        else:
            if self[i] < max(self[lc], self[rc]):
                if self[lc] > self[rc]:
                    self[i], self[lc] = self[lc], self[i]
                    self.shiftdown(lc, end)
                else:
                    self[i], self[rc] = self[rc], self[i]
                    self.shiftdown(rc, end)
        return
    
    def HeapSort(self):
        self.heapify()
        end = len(self)-1
        
        while end>0:
            self[0], self[end] = self[end], self[0]
            end -= 1
            self.shiftdown(0, end)
        return
    
    def QuickSort(self, first, last):
        if first >= last:
            return
        
        pivot = self[first]
        left, right = first+1, last
        while left < right:
            while left < right and self[left] < pivot:
                left += 1
            while right > left and self[right] >= pivot:
                right -= 1
            self[left], self[right] = self[right], self[left]
        
        center = left if self[left]<pivot else left-1
        
        self[first], self[center] = self[center], self[first]
        self.QuickSort(first, center-1)
        self.QuickSort(center+1, last)
    
        return
    
```
